##### S3'den bütün verileri çekmek zor olduğu için localde tuttuğumuz .csv dosyalarını işledik.


```python
import pandas as pd
import os

# Directory path
directory = r'C:\Users\gkhne\Documents\GitHub\end-to-end-video-game-ml\data\raw'

# List to store the dataframes
dfs = []

# Iterate over the files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)
        df = pd.read_csv(file_path)
        dfs.append(df)
        print(f'İşlem Tamamlandı: {filename}')

# Concatenate all dataframes into a single dataframe
result_df = pd.concat(dfs, ignore_index=True)

# Show the result dataframe (optional)
result_df.head()
```

```python
import boto3
import pandas as pd
from io import BytesIO

# Create an S3 client
s3 = boto3.client('s3', aws_access_key_id=aws_access_key_id, aws_secret_access_key=aws_secret_access_key, region_name='us-west-2')

# Specify the bucket name and file name
bucket_name = 'videogames-rawg-ml'
file_name = 'result.parquet'

# Download the file from S3
response = s3.get_object(Bucket=bucket_name, Key=file_name)

# StreamingBody objesini oku
streaming_body = response['Body']

# BytesIO objesine çevir
data = streaming_body.read()
data_io = BytesIO(data)

# DataFrame'e çevir
df = pd.read_parquet(data_io)

# Display the dataframe
print(df.head())
```
